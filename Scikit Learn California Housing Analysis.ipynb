## Analysis of California Housing Data

```python
# Importing necessary libraries based on the order of use in our code

import sklearn.datasets as ds
import numpy as np
import pandas as pd
import ipykernel.pylab.backend_inline # importing this to ensure matplot pyplots accept figure size in first run of cell
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression as LR
```

```python
# Save the california housing dataset

dataset=ds.fetch_california_housing()
```

```python
# Print Description to get an overview of the dataset

print(dataset['DESCR'])
```

```python
# Create a dataframe for independent variables

california_housing_data=pd.DataFrame(data=dataset['data'],columns=dataset['feature_names'])
```

```python
# Create a dataframe for dependent variables

california_housing_values=pd.DataFrame(data=dataset['target'],columns=['target'])
```

```python
# Print first five rows of the dataframe for independent variables

print(california_housing_data.head())
```

```python
# Print first five rows of the dataframe for dependent variable

print(california_housing_values.head())
```

## Summary Statistics

```python
# Display the summary statistics of attributes

print(california_housing_data.describe())
```

```python
# Check for missing values to understand if we need to pre-process and clean data having NA values

california_housing_data.info()
```

#### Observations

From the above information, we make the below conclusions:

* The dataset contains 20,640 samples and 8 column attributes
* All independent variables are numerical features encoded as floating number type
* There are no missing values present in any of the attributes
    
Therefore, there is no need for pre-processing data to handle null values since data present is free of nulls.

## Visualisation

### Histogram

```python
# Set the global defaults for all plots

sns.set_theme(style="darkgrid")
plt.rcParams["figure.figsize"] = [10,8]
```

```python
# Historgram for four of the eight numeric attributes (independent variables)

plt.subplot(2,2,1)
plt.hist(california_housing_data['MedInc'],rwidth=0.90)
plt.xlabel("Median Income")
plt.ylabel("Count")
plt.title('Histogram showing Median Income')

plt.subplot(2,2,2)
plt.hist(california_housing_data['HouseAge'],rwidth=0.90)
plt.xlabel("House Age")
plt.ylabel("Count")
plt.title('Histogram showing Median House Age')

plt.subplot(2,2,3)
plt.hist(california_housing_data['Latitude'],rwidth=0.90)
plt.xlabel("Latitude")
plt.ylabel("Count")
plt.title('Histogram showing Latitude')

plt.subplot(2,2,4)
plt.hist(california_housing_data['Longitude'],rwidth=0.90)
plt.xlabel("Longitude")
plt.ylabel("Count")
plt.title('Histogram showing Longitude')

# Adjust the overlapping titles
plt.tight_layout()
plt.show()
```

#### Observations

From the above histograms, we make the below conclusions:

* The distribution for Median Income has a long right tail.
* The Median Income distribution is for most parts a normal distribution with the exception of some people earning a high salary as it can be observed from its positive skewness, it resembles a right-skewed distribution.
* The distribution corresponding to House Age is mostly uniform.

We can explore latitude and longitude as part of scatter plot to yield more information on their impact on the Median House Value in the coming section.

```python
# Histogram for remaining four numeric attributes (indepednent variables)

plt.subplot(2,2,1)
plt.hist(california_housing_data['Population'],rwidth=0.90)
plt.xlabel("Population")
plt.ylabel("Count")
plt.title('Histogram showing Population')

plt.subplot(2,2,2)
plt.hist(california_housing_data['AveOccup'],rwidth=0.90)
plt.xlabel("Average Occupants")
plt.ylabel("Count")
plt.title('Histogram showing Average Number of Occupants')

plt.subplot(2,2,3)
plt.hist(california_housing_data['AveRooms'],rwidth=0.90)
plt.xlabel("Average Rooms")
plt.ylabel("Count")
plt.title('Histogram showing Average Number of Rooms')

plt.subplot(2,2,4)
plt.hist(california_housing_data['AveBedrms'],rwidth=0.90)
plt.xlabel("Average Bedrooms")
plt.ylabel("Count")
plt.title('Histogram showing Average Number of Bedrooms')

# Adjust the overlapping titles
plt.tight_layout()
plt.show()
```

#### Observations

From the above histograms, we make the below conclusions:

* The difference between the 75% quartile and maximum value is huge.
* This conclusion is in line with the summary statistics obtained in earlier section where we observed a huge difference between the 75% quartile and maximum numbers.
    
We can further confirm presence of these outliers using the Box and Whisker plot, which will be covered in the following section.

```python
#Display histogram for dependent variable

plt.hist(california_housing_values['target'],rwidth=0.90)
plt.xlabel("Median House Value")
plt.ylabel("Count")
plt.title('Histogram of Median House Value')
plt.show()
```

#### Observations

From the above histogram, we make the below conclusions:

* The target distribution has a long tail similar to that observed in histogram for Median Income.
* There is a threshold effect for houses priced on the higher end since all houses with a Median House value above 5 are valued at 5 which accounts for the increase in frequency in histogram at Median House Value 5.

### Scatter Plot

```python
# Scatter plot for Longitude vs. Latitude with corresponding Median House Value

sns.scatterplot(data=california_housing_data, 
                x="Longitude", 
                y="Latitude",
                size=dataset["target"], 
                hue=dataset["target"],
                palette="flare", 
                alpha=0.8)
plt.legend(title="Median House Value", 
           bbox_to_anchor=(1.05,1),
           loc="upper left")
plt.title("Median House Value depending on their Geographical Location")
plt.show()
```

#### Observations

From the above scatter plot, we make the below conclusions:

* The data points corresponding to the scatter plot show a graphical respresentation of California state.
* The high-valued houses (Med House Value of 4 and 5) are located on the coast where big cities of California are located.

### Box and Whisker Plot

```python
# Set the global defaults for all plots

plt.rcParams["figure.figsize"] = (16,12)
```

```python
# Analyse Box & Whisker plot for our attributes

plt.subplot(2,4,1)
plt.boxplot(california_housing_data['MedInc'])
plt.title('Median Income Boxplot')

plt.subplot(2,4,2)
plt.boxplot(california_housing_data['HouseAge'])
plt.title('House Age Boxplot')

plt.subplot(2,4,3)
plt.boxplot(california_housing_data['Latitude'])
plt.title('Latitude Boxplot')

plt.subplot(2,4,4)
plt.boxplot(california_housing_data['Longitude'])
plt.title('Longitude Boxplot')

plt.subplot(2,4,5)
plt.boxplot(california_housing_data['Population'])
plt.title('Population Boxplot')

plt.subplot(2,4,6)
plt.boxplot(california_housing_data['AveOccup'])
plt.title('Average Occupants Boxplot')

plt.subplot(2,4,7)
plt.boxplot(california_housing_data['AveRooms'])
plt.title('Average Rooms Boxplot')

plt.subplot(2,4,8)
plt.boxplot(california_housing_data['AveBedrms'])
plt.title('Average Bedrooms Boxplot')

plt.show()
```

#### Observations

From the above Box and Whisker plots, we notice that there are extreme values present for 'Population' , 'Average Occupants' , 'Average Rooms' and 'Average Bedrooms'. This was also observed in the Summary Statistics and Histograms covered in previous sections, therefore confirming the presence of outliers for these attributes.

Due to the presence of outliers in these four attributes, they may not accurately indicate their impact on the Median House Value (Target/Dependent variable).

## Linear Regression Model Fit (R Squared Score)

R-Squared (R^2) Score is a statistical measure of fit which indicates the amount of variation of a dependent variable that is explained by the independent variables in a regression model. It signifies the goodness of fit for linear regression models.

```python
# Calculate the R^2 score for Linear Regression Model

# Create a linear regression object
regression=LR()

# Fit independent and dependent dada
regression.fit(california_housing_data,california_housing_values)

# Obtain the R^2 score
print("R^2 Score:" , regression.score(california_housing_data,california_housing_values))
```

#### Observations

From the R^2 score of 0.61, it indicates some level of correlation between the independent and dependent variables. Since the score is above 0.5, it is worth investigating the dataset and its implications.
